# Data distribution

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.uniform(-10, 10, 10000)
plt.hist(data, 50)
plt.show()
```

# Normal distribution, visualize the probability density function

```python
from scipy.stats import norm

x = np.arange(-3,3,0.001)

#pdf = Power law/exponential PDF
plt.plot(x, norm.pdf(x), c="y")
plt.show()
```

```python
# desired mean
desiredMean = 5.0

# Standard deviation
sd = 2.0

data = np.random.normal(desiredMean, sd, 10000)
plt.hist(data, 50)
plt.show()

```

# Binomial Probability Mass Function

```python
from scipy.stats import binom

n, p = 10, 0.5
data = np.arange(0,10,0.001)
plt.plot(data, binom.pmf(data, n, p))
plt.show()
```

# Poisson Probability Mass Function

Example: My website gets on average 500 visits per day, whats the odds of getting 550?

```python
from scipy.stats import poisson

mean = 500
data = np.arange(400,600, 0.5)
plt.plot(data, poisson.pmf(data, mean), c ="g")
plt.show()
```

```python
print("10% of the data lands at: ", np.percentile(data, 10))
print("20% of the data lands at: ", np.percentile(data, 20))
print("50% of the data lands at: ", np.percentile(data, 50))
print("90% of the data lands at: ", np.percentile(data, 90))
print("99% of the data lands at: ", np.percentile(data, 99))
```
