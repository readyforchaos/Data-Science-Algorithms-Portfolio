```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Scenario: Car (sell) price/age

# Generate data.
ages = np.random.normal(5, 2, 100)
noise = np.random.normal(0, 0.5, 100)
prices = 10000 - (ages + noise) * 1000

# print(ages, prices)

# Plot data.
plt.scatter(ages, prices)
plt.show()

# Get regression line.
slope, intercept, r_value, p_value, std_err = stats.linregress(ages, prices)

# Plot data and fitted line.
def predict(x): 
    return slope * x + intercept

fitline = predict(ages)

plt.scatter(ages, prices)
plt.plot(ages, fitline, c='r')
plt.show()

# Print values.
print(r_value ** 2)
print(p_value)

# Predict age of 8
test = predict(8)
print(test)

```
