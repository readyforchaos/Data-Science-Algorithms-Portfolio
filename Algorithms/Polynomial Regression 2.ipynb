```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score

# Read data.
df = pd.read_excel('boston.xls')

# Select columns.
df2 = df[['LSTAT', 'MV']]

# Create polynomial line.
p3 = np.poly1d(np.polyfit(df2.LSTAT, df2.MV, 4))
xp = np.linspace(0, 40, 100)

# Plot data.
plt.scatter(df2.LSTAT, df2.MV)
plt.plot(xp, p3(xp), c='r')
plt.show()

# Score model.
r2 = r2_score(df2.MV, p3(df2.LSTAT))
print(r2)

# Predict.
print(p3(15))
```
