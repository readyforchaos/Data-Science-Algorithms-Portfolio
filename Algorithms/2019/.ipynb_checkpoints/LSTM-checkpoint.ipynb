```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn import preprocessing
import os, sys, codecs
```

```python
SCALE_TRAIN = True
WINDOW_SIZE = 256
STRIDE_SIZE = 1
MIN_RUN_LENGTH = WINDOW_SIZE * 10
```

```python
SPA_DATA_FILE = 'data/SPA_Nengine_NTC_P3_T3_excitations.csv'
LABELS_FILE = 'data/SPA_deflexion_filtered (1).csv'

feature_df = pd.read_csv(SPA_DATA_FILE)
label_df = pd.read_csv(LABELS_FILE)
feature_df.columns = ["Time", "EngineRPM", "TurboRPM", "Pressure", "Temperature"]
label_df.columns = ["Time", "Deflection"]
df = feature_df.merge(label_df, on="Time")
len(df)
```

```python
from sklearn.model_selection import train_test_split

label_col = "Deflection"

X_train, X_test, y_train, y_test = train_test_split(df[feature_df.columns], 
                                                    df[label_col], test_size=0.3, 
                                                    shuffle=False, random_state=5)
```

```python
# Normalize train and test data using Z-score scaling - Fit normalizer on train and apply to test
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

if SCALE_TRAIN:
    print('Fit scaler on training data')
    print(scaler.fit(X_train))
    print('Mean values of normalized features')
    print(scaler.mean_)
    print('Variance values of normalized features')
    print(scaler.var_)
    print('Apply scaler to training data')
    X_train = scaler.transform(X_train)
    print('Apply scaler to testing data')
    X_test = scaler.transform(X_test)
```

```python
# Verify that data is normalized
print('New mean and variance of training data')
print(np.mean(X_train, axis=0))
print(np.std(X_train, axis=0))
print('New mean and variance of testing data')
print(np.mean(X_test, axis=0))
print(np.std(X_test, axis=0))
```

```python
def sliding_window(X, stepSize, windowSize):
    # slide a window across the array
    for i in range(0, X.shape[0], stepSize):
        # yield the current window
        window = X[i:i + windowSize]
        if window.shape[0] == windowSize:
            yield window

def sliding_label(y, stepSize, windowSize):
    # slide a window across the array
    for i in range(0, y.shape[0], stepSize):
        label  = np.unique(y[i:i + windowSize])
        yield label[0]
```

```python
# Slice the training data in overlapping windows of size WINDOW_SIZE
print('Slicing training data into windows of size %d' % WINDOW_SIZE)
print('Before training data shape: ', X_train.shape, y_train.shape)
X_train = np.stack(sliding_window(X_train, STRIDE_SIZE, WINDOW_SIZE), axis=0)
y_train = list(sliding_label(y_train, STRIDE_SIZE, WINDOW_SIZE))[:X_train.shape[0]]
print('After training data shape: ', X_train.shape, len(y_train))
```

```python
# Slice the testing data in non-overlapping windows of size WINDOW_SIZE
print('Slicing testing data into windows of size %d' % WINDOW_SIZE)
print('Before testing data shape: ', X_test.shape, y_test.shape)
X_test = np.stack(sliding_window(X_test, STRIDE_SIZE, WINDOW_SIZE), axis=0)
y_test = list(sliding_label(y_test, STRIDE_SIZE, WINDOW_SIZE))[:X_test.shape[0]]
print('After testing data shape: ', X_test.shape, len(y_test))
```

```python
###################
#Hyperparameters
###################
seq_length  = X_train.shape[1]    # should be same as WINDOW_SIZE
nb_features = X_train.shape[2]
# batch_size  = 64
learn_rate  = 0.1 #used in ADAM
reg_param   = 1e-7 #regularization
lstm_units  = 128
```

```python
#os.environ['KERAS_BACKEND'] = 'tensorflow'
os.environ['KERAS_BACKEND'] = 'tensorflow'
from keras.preprocessing.sequence import pad_sequences
from keras.utils.np_utils import to_categorical
from keras.optimizers import SGD, Adam, Adagrad

from keras.layers import Embedding
from keras.layers import Dense, Input, Flatten
from keras.layers import Conv1D, AveragePooling1D, GlobalMaxPooling1D, MaxPooling1D
from keras.layers import Activation, Embedding, Lambda, Merge, Reshape, Dropout, LSTM, GRU, Bidirectional, TimeDistributed
from keras.models import Sequential, Model, load_model

from keras import backend as K
from keras.engine.topology import Layer, InputSpec
from keras import initializers, regularizers, optimizers
from keras.callbacks import History, CSVLogger, ModelCheckpoint, EarlyStopping

l2_reg = regularizers.l2(reg_param)
sdg_optimizer  = SGD(lr=learn_rate, decay=0.01, momentum = 0.1, nesterov=True)
adam_optimizer = Adam(lr=learn_rate, clipnorm=1., beta_1=0.9, beta_2=0.999,  epsilon=None, decay=0.99) #amsgrad=True

model = Sequential()
model.add(LSTM(lstm_units, input_shape=(seq_length, nb_features), kernel_regularizer=l2_reg))
model.add(Dense(1, kernel_regularizer=l2_reg))
model.compile(loss='mae', optimizer = 'rmsprop', metrics=['mae'])

print(model.summary())
```

## Save the model

```python
MODEL_FILENAME = '20-08-2018-lstm-64-adam-win256.h5'
```

```python
# serialize model to JSON
model_json = model.to_json()
with open(MODEL_FILENAME, "w") as json_file:
    json_file.write(model_json)
```

## Train

```python
early_stopping = EarlyStopping(monitor='val_mean_absolute_error', min_delta=0.2, 
                               patience=10, verbose=0, mode='max')

# checkpoint
filepath="weights-improvement-{epoch:02d}-{val_mean_absolute_error:.2f}.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_mean_absolute_error', verbose=1, 
                             save_best_only=True, mode='min')

history = model.fit(X_train, np.array(y_train), callbacks = [checkpoint], epochs=30, batch_size=1024, 
                    validation_data=(X_test, np.array(y_test)))
```

```python
def show_history(history):
    # list all data in history
    print(history.history.keys())

    # summarize history for accuracy
    plt.plot(history.history['mean_absolute_error'])
    plt.plot(history.history['val_mean_absolute_error'])
    plt.title('MAE')
    plt.ylabel('mae')
    plt.xlabel('epoch')
    plt.legend(['train', 'valid'], loc='upper left')
    plt.show()

    # summarize history for loss
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'valid'], loc='upper left')
    plt.show()
    
    return

show_history(history)
```

## Evaluation

```python
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score
from sklearn.metrics import roc_curve, log_loss, f1_score, confusion_matrix, classification_report
from time import time
from datetime import datetime

def evaluate(model, test_x, test_y, prob_thresh=0.5, prob=True):
    print(str(datetime.now()))
    t1    = time()
    preds = model.predict(test_x)
    t2    = time()
    print(str(datetime.now()))
    tdiff = t2 - t1
    print('Evaluation time for %d comments = %.6f secs --> %.6f sec/comment\n' % 
                                (len(preds), tdiff, tdiff/len(preds)))
    return preds
```

```python
predsTest = evaluate(loaded_model, X_test, np.array(y_test), prob=True)
```
