```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Get data.
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Get every type of species (uniquely).
iris['species'].unique()
```

```python
# Create a PairGrid. This is good for when you want to map which type of 
# plots that you want to do.
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
# Get data.
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Create a FacetGrid. This allows you to specify the features in columns/rows and spread the
# types of plots that you want to do. You can also map the plot type.
g = sns.FacetGrid(data=tips,col='time',row='smoker')
#g.map(sns.distplot, 'total_bill')
g.map(plt.scatter, 'total_bill', 'tip')
```
