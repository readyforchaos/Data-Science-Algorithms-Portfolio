```python
import seaborn as sns
%matplotlib inline

# Get data.
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Create a linear model plot, using x as the total_bill and y as tip. Hue splits the
# data between a category. Markers define the point shapes.
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'],
          scatter_kws={'s':100})
```

```python
# You can use col and row to create more plots.
#sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='time', hue='sex')
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex',
          aspect=0.6, size=8)
```
