```python
# Import statements.
import seaborn as sns
%matplotlib inline

# Get data.
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Get histogram and KDE (kernal density estimation).
# Visualizapsdfouias distribution.
sns.distplot(tips['total_bill'],kde=True,bins=30)
```

```python
# Joint plot. Shows distribution for multiple features.
# Use hex to show distribution within coordinates.
# Use reg to show linear regression.
# Use kde to show density. Discarding histogram. 
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
# Make a plot for every possible numerical combination in the data frame.
# Hue for categorical feature to color by in graphs.
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
# Rugpot is like a frequency.
sns.rugplot(tips['total_bill'])
```
