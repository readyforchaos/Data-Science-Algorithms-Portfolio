```python
import seaborn as sns
import numpy as np 
%matplotlib inline

# Get data.
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Create a bar plot of category/numerical values. Defaults to mean.
# Use estimator to specify function.
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
# Create a count plot. Same as bar plot but defaults to count estimator.
sns.countplot(x='sex', data=tips)
```

```python
# Create a box plot. Points outside whiskers are outliers. 25% is within the first half. 
# Good for CXO's. Easy way to interpret data.

#sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# Create a violin plot. This illustrates the KDE of the total_pill per day.
# Duplicated on each side. This is good for data scientists.
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
# Create a strip plot. Use jitter to add noise so we can see (point seperation) 
# and determine where there are many points stacked onto eachother.
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex', split=True)
```

```python
# Create a swarm plot. This combines the strip plot and violin plot. Does
# not scale well.

#sns.swarmplot(x='day',y='total_bill',data=tips)

sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips, color='black')
```

```python
# Create a FactorPlot. This is a shortcut for creating above types of plots
# just using the kind property. 
# TODO: What is the black thing?
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
